```python
import re

def clean_prompts(example):
    example["text"] = re.sub('[\<\>\¬ª\|\üïî\‚óÇ\‚ñ∏\‡∑¥]', '', example["text"])
    example["text"] = re.sub('[a-zA-Z]+', '', example["text"])
    example["text"] = re.sub('[\t\n]', ' ', example["text"])
    example["text"] = re.sub('[\,\/\?\.\!\-\;\:\"\‚Äú\%\‚Äò\‚Äù\ÔøΩ\_]{2,}', ' ', example["text"])
    example["text"] = re.sub('[\s]{2,}', ' ', example["text"])
    return example
```

```python
small_dataset = small_dataset.map(clean_prompts, num_proc=10)
```
